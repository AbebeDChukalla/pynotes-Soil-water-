```python
# Function that downloads binary files from PRISM website
# Load modules
import os, sys
import datetime
import zipfile
import time
import numpy as np
from urllib.parse import urljoin
```

```python
def webread(clim_var, retrieve_date, output_dir):
    # Retrieve 4km PRISM grids from the web
    # Example: webread('ppt', '20170101', '/media/andrespatrignani/HDD 1/PRISM_test_PPT')
    
    # Navigate to directory
    clim_var = 'ppt'
    os.chdir(output_dir);

    # Get date of retrieval
    date_format = '%Y%m%d'
    date = datetime.datetime.now().date() - datetime.timedelta(1) # Current day
    retrieve_date = date.strftime(date_format)

    # Read binary file from PRSIM website
    filesep = '/'
    wget_command = 'wget '
    base_url = 'http://services.nacse.org/prism/data/public/4km'
    full_url = urljoin(base_url, clim_var, retrieve_date)
    output_flag = ' -O '
    extension = '.zip'
    filename = 'PRISM_' + clim_var + '_stable_4kmD2_' + retrieve_date + '_bil' + extension
    os.system(wget_command + full_url + output_flag + filename)

    # Decompress file
    foldername = 'PRISM_' + clim_var + '_stable_4kmD2_' + retrieve_date + '_bil'
    zip_prism = zipfile.ZipFile(os.path.join(output_dir, filename))
    zip_prism.extractall(os.path.join(output_dir, foldername))
    zip_prism.close()
    os.system('rm -r ' + filename);
    time.sleep(0.5)
```
