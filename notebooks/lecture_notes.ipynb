# Reproducible research

## What is reproducible research
* Ability to re-create results, tables, figures, and document analyses.
* Data --> figure
* Publish data, code, and figures
* Aggregate steps into a single environment: Python, Matlab, R
* Traditional research flow
    Retrieve data from a website or collect field/ lab data (only step that cannot be
    Store data: MS Excel (in reproducible research we try to store
    Analyze tabular data: Excel, SigmaPlot
    Analyze spatial data: ArcGIS
    Statistical analysis: SAS
    Image analysis: Sigma Scan
    Insert citations: EndNote, Mendeley
    Write manuscript: MS Word, LaTeX
* Then we need to implement corrections and we need to repeat all the steps again. Programming languages offer a blueprint
* Takes more time at the beginning, but saves time in the long run, it becomes reproducible, and documents the steps.
